```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("top250-00-19.csv")

df.head()

```

```python
df.info()
```

```python
df.describe()
```

```python
df['Market_value'] = df['Market_value'].fillna(df['Market_value'].mean())
```

```python
df.head()
```

```python
fig = px.scatter(x=df['Age'],y=df['Transfer_fee'],labels=dict(y="Transfer Fees",x = "Age"))
fig.show()
```

Players with age ranging from (22 to 27) have higher transfer fees

```python
fig = px.scatter(x=df['Position'],y=df['Transfer_fee'],labels=dict(y="Transfer Fees",x = "Player Position"))
fig.show()
```

Right Winger and Centre Forward positioned players had the highest transfer fees.

```python
fig = px.scatter(df, x='Market_value', y='Transfer_fee')
fig.show()
```

Transfer fees for players are higher than their market value.

```python
fig = px.bar(df, x=df['League_to'].value_counts().keys(), y=df['League_to'].value_counts(), title='Leagues with most transfers',labels=dict(y="Number of Players Transfered to",x = "Leagues"))
fig.show()
```

The Premier league had the most players transfered to it with over 1200 players followed by Serie A with 730 and Laliga with 500.

```python
fig = px.bar(df, x=df['Team_to'].value_counts().keys(), y=df['Team_to'].value_counts(), title='Teams with most transfers',labels=dict(y="Number of Players Transfered to",x = "Teams"))
fig.show()
```

The Inter team had the most players transfered to it followed by Chelsea and Man City.

```python
fig = px.bar(df, x=df['Season'].value_counts().keys(), y=df['Season'].value_counts(), title='Teams with most transfers',labels=dict(y="Number of Players Transfered to",x = "Teams"))
fig.show()
```

All seasons had a similar amount of transfers.
