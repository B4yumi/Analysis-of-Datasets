```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python

sns.relplot(
    data=df,
    x="age", y="trtbps",
    hue="sex")

# as seen in the analysis chart below an increase in age is associated with an increase in resting blood pressure
# with women seen with higher blood pressure values than men.
```

```python
sns.relplot(
    data=df,
    x="age", y="thalachh",
    hue="sex")
    # as seen with the chart below an increase in age is associated with a decline in
    # maximum heart rate (as someone gets older the lower maximum heart rate they have)
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="cp", y="trtbps",
    
)
# as seen in the analysis chart below people with asymptomatic chest pain type (3)
#  had the highest resting blood pressure
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="restecg", y="thalachh",
    
)
# as seen in the analysis chart below people (Value 2) that are showing probable 
# resting electrocardiograph abonormality had a much lower max heart rate than normal(Value 0) or elevated heart (Value 1)
```

```python
sns.set_theme(style="ticks",)
sns.boxplot(x="exng", y="thalachh",
            hue="restecg",
            data=df)
sns.despine(offset=10, trim=True)

# the following analysis chart compared maximum heart rate with exercise induced angina and electrocardiograph results
# as seen people with an exercise induced angina had an abnormal (rest ecg : Value 2) and a much lower max heart rate than any other
```
